```python
import pickle

top_products = pickle.load(open("../../src/evaluation/top_products.pkl", "rb"))
asins = set(top_products["101-1000"])
```

```python
import gzip
metadata_path = 'meta_Electronics.json.gz'
metadata = []
with gzip.open(metadata_path) as metadata_file:
    for i, line in enumerate(metadata_file):
        if i % 50000 == 0:
            print(i)
        line = eval(line)
        if line['asin'] in asins:
            metadata.append(line)
        
len(metadata)
```

```python
from collections import defaultdict

# dict(category -> {asin, title, buy_after_viewing})
category_map = defaultdict(list)

for product in metadata:
    #print(product)
    category = product["categories"][0][-1]
    info = dict()
    info["asin"] = product["asin"]
    if "title" in product:
        info["title"] = product["title"]
    if "related" in product and "buy_after_viewing" in product["related"]:
        info["buy_after_viewing"] = product["related"]["buy_after_viewing"][0]
    if category not in category_map:
        category_map[category] = {}
    category_map[category][info['asin']] = info
```

```python
for category, product_map in category_map.items():
    print("{}: {}".format(category, len(product_map)))
    for asin, product in product_map.items():
        title = product["title"] if "title" in product else "??"
        also_bought = product["buy_after_viewing"] if "buy_after_viewing" in product else "None"
        print("{}/{}, compare to: {}\n".format(asin, title, also_bought))
```

```python
for category, metadatas in sorted(category_map.items(), key=lambda x:x[0]):
    print(category, len(metadatas))
    
```

```python
#categories = ['PC Headsets', 'Monitors', 'Camera', 'Webcams', 'Speaker Systems', 'Remote Home Monitoring Systems', 'Point & Shoot Digital Cameras']
#categories = ['PC Headsets', 'Monitors', 'Webcams', 'Point & Shoot Digital Cameras', 'Laptops', 'LED TVs']
categories = ['Monitors']
category_asins = {}
for category in categories:
    metadatas_map = category_map[category]
    asins = [metadata['asin'] for metadata in metadatas_map.values()]
    category_asins[category] = asins
```

```python
import subprocess
def process_asin(asin):
    bash_command = 'python3 double-prop.py {}'.format(asin)
    process = subprocess.Popen(bash_command.split(), stdout=subprocess.PIPE)
    return process.communicate()
```

```python
import json
def grab_results(asin):
    OUTPUT_DIR = '../output/'
    product_qualities_file = 'product_qualities/{}.json'.format(asin)
    snippets_file = 'snippets/{}.json'.format(asin)
    
    product_qualities = json.load(open(OUTPUT_DIR + product_qualities_file, 'r'))
    snippets = json.load(open(OUTPUT_DIR + snippets_file, 'r'))
    
    return product_qualities, snippets 
```

```python
model_output = {}
print(category_asins.keys())
```

```python
ASINS_LIMIT = 8
for category, category_asins_ in category_asins.items():
    print('Processing category {}...'.format(category))
    for asin in category_asins_[:ASINS_LIMIT]:
        print('Processing asin {}...'.format(asin))
        process_asin(asin)
        product_qualities, snippets = grab_results(asin)
        model_output[asin] = {}
        model_output[asin]['product_qualities'] = product_qualities
        model_output[asin]['snippets'] = snippets
        model_output[asin]['metadata'] = category_map[category][asin]
```
